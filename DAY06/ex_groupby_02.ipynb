### [그룹화 처리]
- 주제 : 타이타닉에서 선실에 등급에 따른 생존 비율을 분석

```python
# [1] 모듈 로딩
import pandas as pd
```

```python
# [2] 데이터 준비
file_name='titanic.csv'
```

```python
# [3] CSV---> DF
DF=pd.read_csv(file_name)
DF
```

```python
DF1=DF[DF['survived'] ==1]
DF1
```

```python
# 데이터를 pclass를 기준으로 데이터를 그룹으로 묶어주기
DF2=DF1.groupby('pclass')
DF2
```

```python
DF3=DF2.count()
```

```python
DF4=DF3.loc[:,'survived']
DF4
```

```python
DF4.loc[1]
DF4.loc[2]
DF4.loc[3]
```
