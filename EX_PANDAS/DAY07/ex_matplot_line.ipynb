### [ Matplotlib 활용한 시각화 ]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# [1] 모듈 로딩
# -      패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(y축 데이터) : 기본값
plt.plot([1,2,3,4])
plt.xlabel('DATA') # 애는 Text로 반환값 있음
plt.ylabel('RESULT') # 애도 Text로 반환값 있음
plt.show() # 최종 결과물로 보여주겠다. show!
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(X축 데이터, Y축 데이터)
plt.plot([10,20,30],[5,10,15])
#            X          Y
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(X축 데이터, Y축 데이터, .fmf포맷스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,10,15],'r--')
#            X          Y
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(X축 데이터, Y축 데이터, .fmf포맷스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30],[5,10,15],'k<--')
#            X          Y
#                             k ,<, --
#                            색상 , 마커모양, 선
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(X축 데이터, Y축 데이터, line2D속성으로 설정)
# fmf 포맷스트링 ==> Line2D의 속성으로 설정

# color='red',marker='o',linestyle='dashed',
#         markersize=10,linewidth=3
# https://matplotlib.org/stable/api/pyplot_summary.html에 들어가서 속성별로 값 찾을 수 있어야 함.

# - markeredgecolor => mec = 'blue'
# - markeredgewidth => mew = 5


plt.plot([10,20,30],[5,10,15],
          color='red',marker='o',linestyle='dashed',
          markersize=10,linewidth=3,
          mec = 'blue',mew=5)
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plt.plot([10,20,30],[15,25,35],'bo--')
# plt.plot([10,20,30],[18,29,25],'ro--')
# plt.plot([10,20,30],[25,25,25],'k^-')
plt.plot([10,20,30],[15,25,35],'bo--',[10,20,30],[18,29,25],'ro--',[10,20,30],[25,25,25],'k^-')
plt.show()
```

```python
# [3] 그래프의 부가적인 기능 추가
# - 그래프 제목
# - X축, Y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성값을 범례에서 사용함 (범례주는 방법 1)
plt.plot([10,20,30],[15,25,20],'bo--', label="V")
plt.plot([10,20,30],[5,30,28],'ro--', label="2")
plt.title('[MPLIB LINE2D]') # 제목 달아주는 거
plt.xlabel('DATA') # x축
plt.ylabel('RESULT') # y축
plt.legend()
# plt.legend('values') # 범례주는 방법 2
plt.show()
```

```python
# [3] 그래프의 부가적인 기능 추가
# - 그래프 제목
# - X축, Y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성값을 범례에서 사용함 (범례주는 방법 1)
#                            loc속성으로 위치 지정 가능 [기본:best]
plt.plot([10,20,30],[15,25,20],'bo--', label="V")
plt.plot([10,20,30],[5,30,28],'ro--', label="2")
plt.title('[MPLIB LINE2D]') # 제목 달아주는 거
plt.xlabel('DATA') # x축
plt.ylabel('RESULT') # y축
plt.legend(loc='center right')
plt.show()
```
