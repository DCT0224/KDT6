### [  파이 그래프  ]
- 컬럼(변수)의 데이터의 분포 비율을 시각화

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2] 데이터 준비
F_DATA=r'C:\Users\KDP-14\Desktop\EX_PANDAS6\DAY04\auto_mpg.csv'
```

```python
# [3] CSV => DataFrame으로 저장
dataDF=pd.read_csv(F_DATA)
dataDF
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율
originSR=round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
x=originSR.index.to_list() # 시리즈라서 to_list()
h=originSR.values.tolist() # array라서 tolist()
```

```python
# - 자동차 생산국가에 대한 비율 시각화
# plt.bar(dataDF['origin'].unique(),dataDF['origin'].value_counts())
originBar=plt.bar(x,h,color=['red','yellow','green'],edgecolor=['black'])

plt.xticks([1,2,3],['USA','EU','JP'],rotation=-45,fontsize=15)
#    눈금이 찍히는 위치값/라벨이름

plt.bar_label(originBar,h,label_type='center')
# 막대에 라벨을 설정

plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['USA','EU','JP'],autopct='%.1f%%')
plt.show()
```
