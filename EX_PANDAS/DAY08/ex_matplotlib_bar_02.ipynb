- [3] 데이터 전처리 
    * 기본확인 파악된 내용들 처리

```python
# [3-1] 불필요 데이터 처리
copyDF=dataDF.drop(index=0)
print(f'{copyDF.head(3)}') # 0번행을 없애서 인덱스가 1부터 시작함. 이것도 처리해줘야됨.
```

```python
# [3-2] 결측치 데이터 처리 => isna()/isnull()
copyDF.isna().sum()
```

```python
# - '1984','1985','1986' 컬럼에서 결측치 데이터만 추출
print(copyDF.columns)
a=copyDF[['1984', '1985','1986']].isna()
a.sum(axis=1)>0
a[a.sum(axis=1)>0]
```

```python
# - '전출지별'
copyDF['전출지별']=copyDF['전출지별'].ffill()

# - 수치 데이터 결측치 처리
copyDF.fillna({0:0},inplace=True)
```

```python
# [3-3] 이상치 데이터 처리
copyDF.replace('-',0,inplace=True)
```

```python
# [3-4] 자료형 변환
d_ = dict(zip(copyDF.columns[2:].to_list(), ['int']*(copyDF.shape[1]-2)))
#                                                     앞에 2컬럼 날려야됨.
# copyDF=copyDF.astype({1970:'int',1971:'int',.....,2017:'int'})
print(d_)
copyDF=copyDF.astype(d_)
copyDF
```

* [4] 데이터 가공
    * 대구에서 전지역 전입 인구수 분석
    * 전출지별 - '대구광역시'
    * 전입지별 - '대구광역시','전국' 제외

```python
# - 대구광역시 데이터만 추출
mask =(copyDF['전출지별'] == '대구광역시')
cityDF=copyDF[mask]
cityDF
```

```python
# - 전출지별 컬럼 삭제
cityDF=cityDF.drop('전출지별',axis=1)

# - 전입지별 컬럼 행 인덱스 설정
cityDF.set_index('전입지별',inplace=True)
```

```python
# - 대구광역시, 전국 행 삭제
cityDF.drop(['대구광역시','전국'],inplace=True)
```

* [5] 히스토그램으로 특정 컬럼 데이터 분포 시각화

```python
# - 2000년도 데이터의 분포
cityDF.columns # 컬럼명 확인
```

```python
# 값 기준 정렬 ==> [기본] 오름차순
cityDF['2000'].sort_values(ascending=False)
```

```python
# 2000년도 대구지역에서 타지역으로 이동인구수 분포
plt.hist(cityDF['2000'],bins=15,histtype='stepfilled')
plt.hist(cityDF['2017'],bins=15,histtype='step')
plt.show()
```

```python
fig=plt.figure(figsize=(10,4))
axes = fig.subplots(2,2,sharey=True)
a = [['2000', '2001'],
     ['2002', '2003']]
for i in range(len(axes)):
    for j in range(len(axes[0])):
        axes[i][j].hist(cityDF[a[i][j]],bins=15,histtype='stepfilled')
plt.show()
```

```python
n,bin,obj=plt.hist(cityDF['2000'],bins=15,histtype='stepfilled')
# n: 반환값 , bin : 구간, obj : 그래프
print(f'n => {n}')
print(f'bin => {bin.tolist()}')
labels=list(map(int,bin.tolist()))
plt.xticks(bin[::2],labels=labels[::2],rotation=45)

plt.show()
```
