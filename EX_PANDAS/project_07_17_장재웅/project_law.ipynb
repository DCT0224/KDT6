```python
law_df.columns=('가해자법규위반별(1)','가해자법규위반별(2)',
                 '주야별','월별',
                 '2018.사고건수','2018.사망자수','2018.부상자수',
                 '2019.사고건수','2019.사망자수','2019.부상자수',
                 '2020.사고건수','2020.사망자수','2020.부상자수',
                 '2021.사고건수','2021.사망자수','2021.부상자수',
                 '2022.사고건수','2022.사망자수','2022.부상자수')
```

```python
# 정비 불량의 데이터 확인
law_df[law_df['가해자법규위반별(1)']=='정비불량']
```

```python
# 정비 불량 행들은 유의미한 데이터가 적기 때문에 DF에서 삭제
law_df = (law_df[law_df['가해자법규위반별(1)']=='운전자법규위반'])
law_df
```

```python
# '가해자법규위반별(1)'은 모두 '운전자법규위반'이므로 삭제
# '가해자법규위반별(2)' 컬럼을 '가해자법규위반별'로 바꿈
law_df.drop(labels='가해자법규위반별(1)',axis=1,inplace=True)
law_df.rename({'가해자법규위반별(2)':'가해자법규위반별'},axis=1,inplace=True)
```

```python
law_df
```

```python
# 멀티 인덱스 지정을 통해 데이터 표준화
law_df=law_df.set_index(['가해자법규위반별','주야별','월별'])
```

```python
# 사고 건수만 확인 할 것이기 때문에 년도별 사망자, 부상자 컬럼 삭제
law_df.drop(labels=['2018.사망자수','2018.부상자수',
                    '2019.사망자수','2019.부상자수',
                    '2020.사망자수','2020.부상자수',
                    '2021.사망자수','2021.부상자수',
                    '2022.사망자수','2022.부상자수',
                    ],axis=1,inplace=True)
law_df
```

```python
# 과속 부분에서 자료없음 찾음.
# 찾아보니 2021년부터 과속 데이터가 경찰청으로 넘어감
# 하지만 주,야 별로 나눈 자료가 아니라서 존재하는 데이터로만 진행.
law_df.loc[('과속','주','전체')]
```

```python
law_df.loc[('과속','야','전체')]
```

```python
## for 문 돌리기 위한 사고건수 year 리스트
year_list=['2018.사고건수','2019.사고건수','2020.사고건수','2021.사고건수','2022.사고건수']

## 함수 만들기
def d_sum(x,y):
    total=0
    for i in x:
        total += law_df.loc[(y),i]
    return total
```

```python
## 주간 법규 위반별 사고 건수
anj_sum=d_sum(year_list,('안전운전불이행','주','전체'))
croj_sum=d_sum(year_list,('교차로운행방법위반','주','전체'))
boj_sum=d_sum(year_list,('보행자보호의무위반','주','전체'))
midj_sum=d_sum(year_list,('중앙선침범','주','전체'))
sinj_sum=d_sum(year_list,('신호위반','주','전체'))
sfj_sum=d_sum(year_list,('안전거리미확보','주','전체'))
elj_sum=d_sum(year_list,('기타','주','전체'))
```

```python
## 주간 법규 위반별 사고 건수
any_sum=d_sum(year_list,('안전운전불이행','야','전체'))
croy_sum=d_sum(year_list,('교차로운행방법위반','야','전체'))
boy_sum=d_sum(year_list,('보행자보호의무위반','야','전체'))
midy_sum=d_sum(year_list,('중앙선침범','야','전체'))
siny_sum=d_sum(year_list,('신호위반','야','전체'))
sfy_sum=d_sum(year_list,('안전거리미확보','야','전체'))
ely_sum=d_sum(year_list,('기타','야','전체'))

```

```python
# 2018.주간과속사고건수
spj_2018=law_df.loc[('과속','주','전체'),'2018.사고건수']
```

```python
# 2018.야간과속사고건수
ypj_2018=law_df.loc[('과속','야','전체'),'2018.사고건수']
```

```python
# 2019.주간과속사고건수
spj_2019=law_df.loc[('과속','주','전체'),'2019.사고건수']
```

```python
# 2019.야간과속사고건수
ypj_2019=law_df.loc[('과속','야','전체'),'2019.사고건수']
```

```python
# 2020.주간과속사고건수
spj_2020=law_df.loc[('과속','주','전체'),'2020.사고건수']
```

```python
# 2020.야간과속사고건수
ypj_2020=law_df.loc[('과속','야','전체'),'2020.사고건수']
```

```python
# 주간, 야간 과속 합계
spj_sum=spj_2018+spj_2019+spj_2020
ypj_sum=ypj_2018+ypj_2019+ypj_2020
```

```python
# 주간 합계
j_sum_list=[anj_sum,croj_sum,boj_sum,midj_sum,sinj_sum,sfj_sum,spj_sum,elj_sum]
```

```python
# 야간 합계
y_sum_list=[any_sum,croy_sum,boy_sum,midy_sum,siny_sum,sfy_sum,ypj_sum,ely_sum]
```

```python
col=['안전운전불이행 사고사건','교차로운행방법위반 사고사건','보행자보호의무위반 사고사건','중앙선침범 사고사건','신호위반 사고사건','안전거리미확보 사고사건','과속 사건사고','기타']
idx=['주간','야간']

dataDF=pd.DataFrame([j_sum_list,y_sum_list],index=idx,columns=col)

dataDF.loc['max'] = dataDF.max()
dataDF['주요 요인']=dataDF.idxmax(numeric_only=True,axis=1)
dataDF
```

## 정리
- 야간보다는 주간 교통사고 사건 발생률이 높다. 이는 야간보다는 주간에 사람들의 이동이 더 많기 때문이라고 생각됨.

- 법규위반별 데이터를 보면 안전운전불이행으로 인한 교통 사건사고가 타 데이터에 비해 높은 수치로 나온 것을 알 수 있음.

- 안전운전불이행은 대부분 운전자의 주의 분산으로 인한 전방주시 태만, 졸음 운전이 주를 이룸. 이러한 요인들은 운전자의 주의로 충분히 극복 가능한 부분임
