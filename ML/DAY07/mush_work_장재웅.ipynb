```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
```

```python
## 데이터 준비
FILE_NAME = r'C:\VSCode\KDT\머신러닝\DAY07\mushrooms.csv'
MushDF = pd.read_csv(FILE_NAME)
```

```python
MushDF
```

```python
MushDF.info()
# targetSR => class
```

```python
MushDF['stalk-root'].replace('?','T',inplace=True)
```

```python
# 결측치 확인
MS_column = MushDF.columns
for column in MS_column:
    print(MushDF[column].value_counts())
```

```python
featuresDF = MushDF[MS_column[1:]]
targetSR = MushDF['class']
# X_features_ohe = pd.get_dummies(column)
```

```python
featuresDF
featuresDF.shape
```

```python
targetSR.shape
```

```python
X_features_ohe = pd.get_dummies(featuresDF,columns=MS_column[1:],drop_first=True)

# get_dummies를 통해 one-hot 인코딩 사용
df = X_features_ohe.applymap(lambda x: 1 if x is True else (0 if x is False else x))

# label 인코딩을 통해 target 인코딩해주기
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le = le.fit(targetSR)
lable_encoded = le.transform(targetSR)
targetSR= pd.Series(lable_encoded)
targetSR
```

```python
# 학습용, 테스트용 데이터 셋 분리하기
X_train,X_test,y_train,y_test = train_test_split(df,
                                                 targetSR,
                                                 stratify=targetSR,
                                                 random_state=10)
```

```python
print(f'X_train : {X_train.shape}, {X_train.ndim}D')
print(f'X_test : {X_test.shape}, {X_test.ndim}D')

print(f'y_train : {y_train.shape}, {y_train.ndim}D')
print(f'y_test : {y_test.shape}, {y_test.ndim}D')
```

```python
X_train
```

```python
# DecisionTree를 이용해 상관관계가 높은 컬럼을 뽑은 후 원-핫 인코딩을 사용
dt_model = DecisionTreeClassifier(random_state=10)
dt_model.fit(X_train,y_train)
```

```python
dt_model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_true = y_test
```

```python
y_test_pred = dt_model.predict(X_test)
```

```python
confusion_matrix(y_true,y_test_pred)
```

```python
# 모델 파라미터 확인 : fit()후에 학습용 데이터셋을 기반으로 설정된 값들
print(f'[dt_model.feature_names_in_] {dt_model.feature_names_in_}')
print(f'[dt_model.feature_importantances_] {dt_model.feature_importances_}')
```

```python
feature_name = pd.DataFrame(dt_model.feature_names_in_)
feature_importance = pd.DataFrame(dt_model.feature_importances_)

# 데이터 프레임 결합하기
feature_import = pd.concat([feature_name,feature_importance],axis=1)
feature_import.columns = ['name','value']
feature_import.set_index(feature_import['name'],inplace=True)
feature_import
```
