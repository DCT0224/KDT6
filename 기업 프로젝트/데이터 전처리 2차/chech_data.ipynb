## 데이터 확인

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## 데이터 불러오기
FILE_PATH = r'C:\Users\KDP-14\Desktop\VSCode\KDT6\기업 프로젝트\Dataset\바웰공정데이터(2차데이터).csv'
dataDF = pd.read_csv(FILE_PATH)
```

```python
dataDF
```

```python
## 사출온도 설정값 = 0 지우기 (48개)
dataDF = dataDF[dataDF['n_temp_sv'] !=0]
```

```python
## 설정값 날려주기
dataDF.drop(columns=['time','E_scr_sv','c_temp_sv','k_rpm_sv','n_temp_sv','s_temp_sv'],inplace=True)
```

```python
dataDF.columns
```

### 데이터별 value_counts 및 hist

```python
# 각 컬럼별 histogram 그려보기
for i in dataDF.columns:
    print(dataDF[i].value_counts())
    print(dataDF[i].describe())
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    # 박스플롯
    sns.boxplot(data=dataDF[i], ax=axes[0], color="skyblue")
    axes[0].set_title("Boxplot")

    # 히스토그램
    sns.histplot(dataDF[i], ax=axes[1], kde=True, color="lightcoral")
    axes[1].set_title("Histogram")

    # 그래프 간격 조정
    plt.tight_layout()
    plt.show()
```

### 데이터 전처리

```python
## E_scr_pv 컬럼 날리기
dataDF.drop(columns='E_scr_pv',inplace=True)
```

```python
## 칼날 rpm 40 초과인 값만 남기기
dataDF = dataDF[dataDF['k_rpm_pv']>40]
```
