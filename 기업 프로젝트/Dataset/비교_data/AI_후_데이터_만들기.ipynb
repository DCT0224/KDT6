```python
## 비교 데이터 셋 만들기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\KDP-14\Desktop\VSCode\KDT6\기업 프로젝트\Dataset\비교_data\total_refill_RF_time_1126.csv'
data = pd.read_csv(file_path)
```

```python
data
```

```python
# loss 컬럼 만들기
data['loss'] = abs(data['scale_pv'] - 3)
```

```python
# quality 컬럼 만들기
data['quality'] = data['scale_pv'].apply(
    lambda x: 'A' if 2.9 <= x <= 3.1 else ('B' if 2.8 <= x <= 3.2 else 'C'))
```

```python
data['datetime_type']= pd.to_datetime(data['time'])
data['date_only'] = data['datetime_type'].dt.date
```

```python
# 'time','datetime_type' 컬럼 없애주기
dataDF = data.drop(columns=['time','datetime_type','Day'])
```
