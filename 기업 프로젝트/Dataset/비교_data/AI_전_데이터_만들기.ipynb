```python
## 비교 데이터 셋 만들기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\KDP-14\Desktop\VSCode\KDT6\기업 프로젝트\Dataset\바웰공정데이터(2차데이터).csv'
data = pd.read_csv(file_path)
```

```python
data['scale_pv'].value_counts().sort_values(ascending=False)
```

```python
## 설정값 날려주기
data.drop(columns=['E_scr_pv','E_scr_sv','c_temp_sv','k_rpm_sv','n_temp_sv','s_temp_sv'],inplace=True)
```

```python
# k_rpm 실제값 = 0 & 40 지움
data = data.drop(data[(data['k_rpm_pv'] == 0) | (data['k_rpm_pv'] == 40)].index)
```

```python
data.reset_index(inplace=True)
data.drop(columns='index',inplace=True)
```

```python
data
```

```python
def change_scale(dataDF, x, y):
    # scale_pv가 x와 y 사이인 값은 유지, 그렇지 않은 값은 0으로 설정
    dataDF['scale_pv'] = dataDF['scale_pv'].apply(lambda value:
                                                  value if float(x) <= value <= float(y) else 0)
    return dataDF
```

```python
change_scale(data,2.7,3.3)
```

```python
data['scale_pv'].value_counts()
```

```python
data['scale_pv'] = data['scale_pv'].apply(lambda value:
                                                  3.11 if value == 0 else value)
```

```python
data['scale_pv'].value_counts()
```

```python
# loss 컬럼 만들기
data['loss'] = abs(data['scale_pv'] - 3)
```

```python
# quality 컬럼 만들기
data['quality'] = data['scale_pv'].apply(
    lambda x: 'A' if 2.9 <= x <= 3.1 else ('B' if 2.8 <= x <= 3.2 else 'C'))
```

```python
data['datetime_type']= pd.to_datetime(data['time'])
data['date_only'] = data['datetime_type'].dt.date
```

```python
# 'time','datetime_type' 컬럼 없애주기
dataDF = data.drop(columns=['time','datetime_type'])
```
